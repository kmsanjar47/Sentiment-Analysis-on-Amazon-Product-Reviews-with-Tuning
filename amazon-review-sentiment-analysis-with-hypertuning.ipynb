# Data Preprocessing

## Null Handling

```python
df.isna().sum()
```

## Text Preprocessing

```python
df["reviewText"][0]
```

```python
stopwords.fileids()
```

```python
stopword = stopwords.words("english")
```

```python
punctuation = string.punctuation
```

```python
def preprocess_text(text):
 
    preprocess_punctuation = "".join([char for char in text if char not in punctuation])
    preprocess_stopword = [char for char in preprocess_punctuation.split() if char.lower() not in stopword]
    return preprocess_stopword
```

```python
df["reviewText"] = df["reviewText"].apply(preprocess_text)
```

```python
df.head()
```

# Text Normalization/Scaling

```python
lemmatizer = WordNetLemmatizer()
```

```python
df["reviewText"][0]
```

```python
def normalize_data(text):
    normalized_text = ' '.join([lemmatizer.lemmatize(word) for word in text])
    return normalized_text
```

```python
import nltk  #Handle Lookup Error
nltk.download("wordnet")
!unzip /usr/share/nltk_data/corpora/wordnet.zip -d /usr/share/nltk_data/corpora/
```

```python
df["reviewText"] = df["reviewText"].apply(normalize_data)
```

```python
df.head()
```

# Vectorization

```python
vectorizer = TfidfVectorizer()
```

```python
X = vectorizer.fit_transform(df["reviewText"])
y = df["Positive"]
```

```python
X.shape
```

# Data Spliting

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
xtrain.shape
```

# Model Training

```python
mnbNB = MultinomialNB()
berNB = BernoulliNB()
lr = LogisticRegression()
```

```python
models = [mnbNB, berNB, lr]
```

```python
def train_models(X_train,y_train, models):
    for model in models:
        model.fit(X_train,y_train)
        print(f"{model} : {mt.accuracy_score(y_train, model.predict(X_train))}")   
```

```python
train_models(xtrain,ytrain,models)
```

# Model Evaluation

```python
def evaluate_model(X_test, y_test, models):

    y_predicted_list = []
    accuracy_scores = []

    for model in models:
        prediction = model.predict(X_test)
        y_predicted_list.append(prediction)
        accuracy_scores.append(mt.accuracy_score(y_test,prediction))
    
    for idx, prediction in enumerate(y_predicted_list):
        print(f"{models[idx]}")
        print()
        print("Test Accuracy Score: ",accuracy_scores[idx])
        print()
        print(mt.classification_report(y_test,prediction))
        print("-------------------------------------------------------------------------")

    return accuracy_scores, y_predicted_list
```

```python
accuracy_scores, predictions = evaluate_model(xtest,ytest,models)
```

<p>By analyzing the classification reports above, we can see that Logistic Regression has the best accuracy with 88%.

# HyperParameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'penalty': ['l1', 'l2'],          # Regularization type (L1 or L2)
    'C': [0.001, 0.01, 0.1, 1, 10],    # Inverse of regularization strength
    'solver': ['liblinear', 'saga']   # Algorithm to use for optimization
}
```

```python
scoring = {
    'accuracy': mt.make_scorer(mt.accuracy_score),
    'precision': mt.make_scorer(mt.precision_score, pos_label='positive'),
    'recall': mt.make_scorer(mt.recall_score, pos_label='positive'),
    'f1': mt.make_scorer(mt.f1_score, pos_label='positive')
}
```

```python
grid_search = GridSearchCV(estimator=lr, param_grid=param_grid, cv=5, scoring=scoring, refit="accuracy")
grid_search.fit(xtrain, ytrain)
```

```python
best_params = grid_search.best_params_
```

```python
lr_best = LogisticRegression(**best_params)
```

```python
lr_best.fit(xtrain,ytrain)
```

```python
predicted = lr_best.predict(xtest)
```

```python
print(mt.classification_report(ytest,predicted))
```
